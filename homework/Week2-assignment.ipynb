````python
text_list_capitalize_round = []

for item in text_list[1:]:
    text_list_capitalize_round.append(
        [item[0].capitalize(), round(float(item[1]), 2), round(float(item[2]), 2)])
                                 
text_list_capitalize_round
````

### Hard Questions

#### 1. Create a function that takes a list of words and returns step 3 till 6 as a list
 - do not forget the comments and to explain what your function does and what it returns (what is the expected function output)

````python
def applying_steps_3_to_6 (text):
    '''This function get a list and apply methods to:
            Make sure all the countries names starts from capital letters
            Round all the numbers to 2 decimals
            Sort the list based on the country names from A -> Z
            Remove the last letter of every country
        Input: list -> [country, birthrates and Life expectancy]
        Output: list -> [country, birthrates and Life expectancy]
    '''
    
    text_list = [line.strip().split('\t') for line in text]
    
    for item in text_list[1:]:
        index = text_list.index(item)
        text_list[index] = [item[0][:-1].capitalize(), round(float(item[1]), 2), round(float(item[2]), 2)]
    
    text_list = sorted(text_list)
    
    return(text_list)    
````

````python
applying_steps_3_to_6(text)
````

#### 2. How many country does the file contain?

````python
countries = []

for item in text_list[1:]:
    countries.append(item[0])

len(set(countries))
````

#### 3. Do all countries have birth and life expectancy present?

Before that, all the numbers was converting to float. So, if there are a number it'll be a float. <br>
I considered the second and third columns as birth and expectancy, respectively. After all, it's what the header show us.

````python
answer = True

for item in text_list[1:]:
    if not (isinstance(item[1], np.float64) 
            and isinstance(item[2], np.float64)):
        answer = False
        break

print(answer)
````

#### 4. How many countries start with the letter M?

All the countries start with uppercase. It was applying in previous steps.

````python
count = 0

for country in countries:
    if country[0] == 'M':
        count += 1
        
print(count)
````

#### 5. How many countries have a 'z' in their name?

````python
count = 0

for country in countries:
    if 'z' in country:
        count += 1
        
print(count)
````
